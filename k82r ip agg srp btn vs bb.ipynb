```python
row_drop = []
for i in range(565,577,1):
    row_drop.append(i)
```

```python
df_new = df.drop(row_drop,axis=0)
```

```python
plt.scatter([range(0,565)],df_new.equity)
plt.show()
```

```python
k82r_data = df_new[['equity','ev','f_bet200','f_bet100','f_bet50','f_bet25','f_check']]
```

```python
k82r_data
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('standardize', StandardScaler(), ['equity','ev','f_bet200','f_bet100','f_bet50','f_bet25','f_check'])], remainder='passthrough')
```

```python
data_preprocessed = ct.fit_transform(k82r_data)
```

```python
data_preprocessed = pd.DataFrame(data_preprocessed)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4,random_state=0)
```

```python
cluster = kmeans.fit(data_preprocessed)
```

```python
cluster.labels_
```

```python
result = pd.DataFrame()
```

```python
result['hands'] = df_new['hands']
```

```python
result['group'] = pd.DataFrame(cluster.labels_)
```

```python
result
```

```python
interested_group = []
```

```python
for index,row in result.iterrows():
    if row['group'] == 0:
        #print(row['hands'])
        print(row['hands']+':'+'1'+',',end='')
        interested_group.append(row['hands'])
print('\b')    
```

```python
avg_equity = 0
avg_ev = 0
c = len(interested_group)
for index,row in df.iterrows():
    for hand in interested_group:
        if (row['hands']==hand):
            avg_equity += row['equity']
            avg_ev += row['ev']
print(avg_equity/c,avg_ev/c)
```
